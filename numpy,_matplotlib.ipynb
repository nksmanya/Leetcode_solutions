```python
import matplotlib.pyplot as plt
overs= np.arange(1,11)
score = [1,6,20,3,5,6,13,5,7,21]
plt.bar (overs,score,color='purple')
plt.title ("India's score over 10 over(Bar Chart)")
plt.xlabel("Over Number")
plt.ylabel("Runs Scored")
plt.xticks(overs)
plt.grid( axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


overs = np.arange(1, 11)
score = [1, 6, 20, 3, 5, 6, 13, 5, 7, 21]

plt.plot(overs, score, marker='o', linestyle='-', color='green', label="Runs per Over")


plt.title("India's Score Over 10 Overs (Line Graph)")
plt.xlabel("Over Number")
plt.ylabel("Runs Scored")


plt.xticks(overs)

plt.grid(True)
plt.legend()


plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random
import time

# Create figure and axes
fig, ax = plt.subplots()
line, = ax.plot([], [], 'o-') # Initialize an empty line plot

# Initialize data lists
x_data, y_data = [], []

def update_plot(frame):
    # Simulate new data arrival
    new_x = time.time() # Use timestamp for x-axis
    new_y = random.randint(1, 100) # Simulate a new data point

    x_data.append(new_x)
    y_data.append(new_y)

    # Keep only the last N data points for a moving window
    max_points = 20
    if len(x_data) > max_points:
        x_data.pop(0)
        y_data.pop(0)

    line.set_data(x_data, y_data)
    ax.relim() # Recalculate limits for the axes
    ax.autoscale_view(True,True,True) # Autoscale the view

    return line,

# Create animation
ani = animation.FuncAnimation(fig, update_plot, interval=1000, blit=True) # Update every 1000ms (1 second)

plt.show()
```
